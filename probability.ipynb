# Комбинаторика

## Размещения
Размещением (из $n$ по $m$) называется **упорядоченный** набор из $m$ различных элементов из некоторого множества различных $n$ элементов. Формула для расчета:
$$A_n^m = \frac{n!}{(n-m)!}$$

## Сочетания
Сочетанием (из $n$ по $m$) называется **неупорядоченный** набор из $m$ различных элементов из некоторого множества различных $n$ элементов. Формула для расчета:
$$C_n^m = \frac{n!}{m!(n-m)!}$$

## Перестановки
Частный случай размещения, когда $m$ совпадает с $n$. Формула для расчета:
$$ P = n! $$

## Размещения с повторениями
$$\tilde{A}_n^m = n^m$$

## Сочетания с повторениями
$$ \tilde{C}_n^m = C_{n+m-1}^m $$

## Перестановки с повторениями
$$ P_n(n_1,n_2,...,n_k) = \frac{n!}{n_1!n_2!...n_k!} $$

# Условная вероятность
Вероятность события $B$, найденное при условии, что событие $A$ произошло, называется условной вероятностью события $B$ и обозначается $P_A(B)$ или $P(B|A)$.

Вероятность произведения двух событий равна произведению вероятности одного из них на условную вероятность другого, найденную в предположении, что первое событие произошло:
$$ P(AB) = P(A)P_A(B) = P(B)P_B(A) $$

Если событие $F$ может произойти только при условии появления одного из событий (гипотез) $A_1,A_2,...,A_n$, образующих полную группу, вероятность события $F$ равна сумме произведений вероятностей каждого из этих событий (гипотез) на соответствующие условные вероятности $F$:
$$ P(F) = \sum_{i=1}^n P(A_i)P_{A_i}(F) $$
Это **формула полной вероятности**.

# Формула Байеса
$$ P_F(A_i) = \frac{P(A_i)P_{A_i}(F)}{P(F)} = \frac{P(A_i)P_{A_i}(F)}{\sum_{i=1}^nP(A_i)P_{A_i}(F)} $$

# Формула Бернулли
Если вероятность $p$ наступления события $A$ в каждом испытании постоянна, то веростноять $P_{m,n}$ того, что событие $A$ наступит $m$ раз в $n$ независимых испытаниях, равна
$$ P_{m,n} = C_n^m p^m q^{n-m}, $$
где $q=1-p$.

# Формула Пуассона
Если вероятность $p$ наступления события $A$ в каждом испытании стремится к нулю ($p \to 0$) при неограниченном увеличении числа $n$ испытаний ($n \to \infty$), при чем произведение $np$ стремится к постоянному числу $\lambda$ ($np \to \lambda$), то вероятность $P_{m,n}$ того, что событие $A$ появится $m$ раз в $n$ независимых испытаниями, удовлетворяет предельному равенству:
$$ \lim\limits_{n \to \infty} P_{m,n} = P_m(\lambda) = \frac{\lambda^m e^{-\lambda}}{m!} $$

# Локальная формула Муавра-Лапласа
Если вероятность $p$ наступления события $A$ в каждом испытании постоянна и отлична от 0 и 1, то вероятность $P_{m,n}$ того, что событие $A$ произойдет $m$ раз в $n$ независимых испытаниях при достаточно большом числе $n$ приближенно равна
$$ P_{m,n} \approx \frac{f(x)}{\sqrt{npq}}, $$
где $f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}$ — функция Гаусса и $x = \frac{m - np}{\sqrt{npq}}$.

# Интегральная формула Муавра-Лапласа
Если вероятность $p$ наступления события $A$ постоянна и отлична от 0 и 1, то вероятность того, что число $m$ наступления события $A$ в $n$ независимых испытаниях заключено в пределах от $a$ до $b$ (включительно), при достаточно большом числе $n$ равна
$$ P_n(a \leq m \leq b) \approx \frac{1}{2} [\Phi(x_2) - \Phi(x_1)], $$
где $\Phi(x) = \frac{2}{\sqrt{2\pi}} \int_0^x e^{-\frac{t^2}{2}}{d}x$ — функция (или интеграл вероятностей) Лапласа; $x_1 = \frac{a - np}{\sqrt{npq}}$, $x_2 = \frac{b - np}{\sqrt{npq}}$.

# Случайные величины

**Случайной величиной $X$** называется функция, заданная на множестве элементарных исходов (или в пространстве элементарных исходов), т.е.
$$ X = f(\omega), $$
где $\omega$ — элементарный исход (или элементарное событие, принадлежащее пространству $\Omega$, т.е. $\omega \in \Omega$).

**Законом распределения** случайной величины называется всякое соотношение, устанавливающее связь между возможными значениями случайной величины и соответствующими им вроятностями.

Дву случайные величины называются *независимыми*, если закон распределения одной из них не меняется от того, какие возможные значения приняла другая величина (то есть события $X = x_i$ и $Y = y_i$ независимы при любых $i = 1,2,...,n$ и $j = 1,2,...,n$).

*Произведением $kX$ случайной величины $X$ на постоянную величину $X$* называется случайная величина, которая принимает значения $kx_i$ с теми же вероятностями $p_i$ ($i = 1,2,...,n$).

*$m$-й степенью случайной величины $X$*, т.е. $X^m$, называется случайная величина, которая принимает значения $x_i^m$ с теми же вероятностями $p_i$ ($i = 1,2,...,n$).

*Суммой (разностью или произведением) случайных величин $X$ и $Y$* называется случайная величина, которая принимает все возможные значения вида $x_i + y_j$ ($x_i - y_j$ или $x_i \cdot y_j$), где $i = 1,2,...,n$; $j = 1,2,...,m$ с вероятностями $p_{ij}$ того, что случайная велечина $X$ примет значение $x_i$, а $Y$ — значение $y_j$.
$$ p_{ij} = P[(X=x_i)(Y=y_j)]. $$

**Математическим ожиданием** $M(X)$ дискретной случайной величины $X$ называется сумма произведений всех ее значений на соответствующие вероятности:
$$ M(X) = \sum_{i=1}^n x_i p_i $$
Математическое ожидание рассматривают как центр распределения случайной величины.

*Свойства математического ожидания:*
1. $M(C) = C,$ где $C$ — постоянная величина.
2. $M(kX) = kM(X),$ где $k$ — постоянный множитель.
3. $M(X \pm Y) = M(X) \pm M(Y)$
4. $M(XY) = M(X)M(Y)$
5. $M(X \pm C) = M(X) \pm C$
6. $M(X - M(X)) = 0$

**Дисперсией** $D(X)$ случайной величины $X$ называется математическое ожидание квадрата ее отклонения от математического ожидания:
$$D(X) = M[X - M(X)]^2 = M(X - a)^2,$$
где $a = M(X)$. 

**Средним квадратическим отклонением (стандартным отклонением)** $\sigma_x$ случайной величины $X$ называется арифметическое значение квадратного корня из ее дисперсии:
$$ \sigma_x = \sqrt{D(X)} $$

*Свойства дисперсии:*
1. $D(C) = 0,$ где $C$ — постоянная величина.
2. $D(kX) = k^2D(X),$ где $k$ — постоянный множитель.
3. $D(X) = M(X^2) - [M(X)]^2 = M(X^2) - a^2,$ где $a = M(X)$.
4. $D(X + Y) = D(X - Y) = D(X) + D(Y)$

**Функцией распределения** случайной величины $X$ называется функция $F(x)$, выражающая для каждого $x$ вероятность того, что случайная величина $X$ примет значение, меньшее $x$:
$$ F(x) = P(X < x) $$
Функцию $F(x)$ иногда называют *интегральной функцией распределения* или *интегральным законом распределения*.

*Свойства функции распределения:*
1. $ 0 \leq F(x) \leq 1 $
2. Функция распределения случайной величины есть неубывающая функция на всей числовой оси.
3. $ F(-\infty) = \lim\limits_{x \to -\infty} F(x) = 0 $, $ F(+\infty) = \lim\limits_{x \to +\infty} F(x) = 1 $
4. $P(x_1 \leq X < x_2) = F(x_2) - F(x_1)$

Случаяная величина $X$ называется **непрерывной**, если ее функция распределения непрерывная в любой точке и дифференцируема всюду, кроме, быть может, отдельных точек. \
**Теорема.** Вероятность любого отдельно взятого значения $x_1$ непрерывной случайной величины $X$ равна нулю:
$$ P(X = x_1) = 0 $$

**Плотностью вероятности (плотностью распределения)** $\phi(x)$ непрерывной случайной величины $X$ называется производная ее функции распределения:
$$ \phi(x) = F'(x) $$

Свойства плотности вероятности:
1. $\phi(x) \geq 0$
2. $P(a \leq X \leq b) = \int\limits_a^b \phi(x) {d}x$
3. $F(x) = \int\limits_{-\infty}^x \phi(x) {d}x$
4. $\int\limits_{-\infty}^{+\infty} \phi(x) {d}x = 1$

Математическое ожидание непрерывной случайной величины:
$$ M(X) = \int\limits_{-\infty}^{+\infty} x\phi(x) {d}x $$
Дисперсия непрерывной случайной величины:
$$ D(X) = \int\limits_{-\infty}^{+\infty} (x - a)^2 \phi(x) {d}x, $$
где $a = M(X)$

**Модой** $Mo(X)$ случайной величины $X$ называется ее наиболее вероятное значение (для которого вероятность $p_i$ или плотность вероятности $\phi(x)$ достигает максимума).

**Медианной** $Me(x)$ непрерывной случайной величины $X$ называется такое ее значение, для которого
$$ P(X < Me(X)) = P(X > Me(x)) = \frac{1}{2} $$

**Квантилем уровня** $q$ ($q$-квантилем) называется такое значение $x_q$ случайно величины, при котором функция распределения принимает значение, равное q, т.е.
$$ P(x_q) = P(X < x_q) = q $$

* $x_{0.5}$ — медиана
* $x_{0.25}$ — нижний квартиль
* $x_{0.75}$ — верхний квартиль

**Начальным моментом** $k$-го порядка случайной величины $X$ называется математическое ожидание $k$-й степени этой величины:
$$ v_k = M(X^k) $$

**Центральным моментом** $k$-го порялка случайной величины $X$ называется математическое ожидание $k$-й степени отклонения случайной величины $X$ от ее математического ожидания:
$$ \mu_k = M[X-M(X)]^k $$

# Основные законы распределения
## Биномиальный закон распределения

Дискретная случайная величина $X$ имеет **биномиальный закон распределения** с параметрами $n$ и $p$, если она принимает значения $0, 1, 2, ..., m, ..., n$ с вероятностями
$$P(X=n) = C_n^m p^m q^{n-m},$$
где $0<p<1$, $q=1-p$.

Математическое ожидание случайной величины $X$, распределенной по биномиальному закону,
$$ M(X) = np, $$
а ее дисперсия
$$ D(X)=npq. $$

## Закон распределения Пуассона

Дискретная случайная величина $X$ имеет **закон распределения Пуассона** с параметром $\lambda > 0$, если она принимает значения $0,1,2,...,m,...$ (бесконечное, но счетное множество значений) с вероятностями
$$ P(X=m) = \frac{\lambda^m e^{-\lambda}}{m!} = P_m(\lambda). $$

Математическое ожидание и дисперсия случайной величины, распределенной по закону Пуассона совпадают и равны параметру $\lambda$ этого закона, т.е.
$$ M(X) = \lambda, \; M(X) = \lambda. $$

## Геометрическое распределение

Дискретная случаяная величина $X = m$ имеет **геометричесове распределение**, если она принимает значения $1,2,...,m,...$ (бесконечное, но счетное множество значений) с вероятностями
$$ P(X=m) = pq^{m-1}, $$
где $0<p<1, \; q=1-p.$

Случайная величина $X=m$, имеющая геомтерическое распределение, представляет собой число $m$ испытаний, проведенных по схеме Бернулли, с вероятностью $p$ наступления события в каждом испытании до первого положительного исхода.

Математическое ожидание случайной величины $X$, имеющей геомтерическое распределение с параметром $p$,
$$ M(X) = \frac{1}{p}, $$
а ее дисперсия
$$ D(X) = \frac{q}{p^2}.  $$

Геометрическое распределение при $k=1$ является частным случаем **распределения Паскаля**, для которого
$$ P(X=m) = C_{m-1}^{k-1} p^k q^{m-k}, \; m=k, k+1,... $$
и числовые характеристики
$$ M(X) = \frac{k}{p}, \; D(X) = \frac{kq}{p^2}. $$

Геометрическое распределение характеризует число $m$ испытаний до первого положительного исхода; *распределение Паскаля — до $k$-го положительного исхода*.

## Гипергеометрическое распределение

Дискретная случаяная величина $X$ имеет **гипергеометрическое распределение** с параметрами $n, \; M, \; N$, если она принимает значения $0,1,2,...,m,...,min(n,M)$ с вероятностями
$$ P(X=m) = \frac{C_M^n C_{N-M}^n-m{}}{C_N^n}, $$
где $M \leq N, \; n \leq N; \; n, M, N$ — натуральные числа.

Гипергеометрическое распределение имеет случайная величина $X=m$ — число объектов, обладающих заданным свойством, среди $n$ объектов, случайно извлеченных (без возврата) из совокупности $N$ объектов, $M$ из которых обладают этим свойством.

Математическое ожидание случайной величины $X$, имеющей гипергеометрическое распределение с параметрами $n, \; M, \; N,$ есть
$$ M(X) = n \frac{M}{N}, $$
а ее дисперсия
$$ D(X) = n \frac{M}{N-1} (1 - \frac{M}{N}) (1 - \frac{n}{N}). $$

## Равномерный закон распределения

Непрерывная случайная величина $X$ имеет равномерный закон распределения на отрезке $[a,b]$, если ее плотность вероятности $\phi$ постоянна на этом отрезке и равна нуля вне его, т.е.
$$
\phi(x) =
  \begin{cases}
    \frac{1}{b-a} \; при \; a \leq x \leq b\\
    0 \; при \; x < a, \; x > b
  \end{cases}
$$

Функция распределения случайной величины, распределенной по равномерному закону, есть
$$
F(x) =
  \begin{cases}
    0 \; при \; x \leq a\\
    \frac{x-a}{b-a} \; при \; a < x \leq b\\
    1 \; при \;  x > b
  \end{cases}
$$
ее математическое ожидание
$$ M(X) = \frac{a+b}{2}, $$
а дисперсия
$$ D(X) = \frac{(b-a)^2}{12}. $$

## Показательный закон распределения

Непрерывная случайная величина $X$ имеет **показательный закон распределения** с параметром $\lambda > 0$, если ее плотность распределения имеет вид:
$$
\phi(x) =
  \begin{cases}
    \lambda e^{-\lambda x} \; при \; x \geq 0\\
    0 \; при \; x < 0
  \end{cases}
$$

Функция распределения случайной величины $X$, распределенной по показательному закону, есть
$$
F(x) =
  \begin{cases}
    0 \; при \; x < 0\\
    1 - e^{-\lambda x} \; при \;  x \geq 0
  \end{cases}
$$
ее математическое ожидание
$$ M(X) = \frac{1}{\lambda} $$
а дисперсия
$$ D(X) = \frac{1}{\lambda^2} $$

## Нормальный закон распределения

Непрерывная случайная величина $X$ имеет **нормальный закон распределения (закон Гаусса)** с параметрами $a$ и $\sigma$, если ее плотность вероятности имеет вид:
$$ \phi_N(x) = \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{(x-a)^2}{2\sigma^2}} $$

![normal distribution density](data/normal_dist_dens.png "Normal distribution density")

Математическое ожидание случайной величины $X$, распределенной по нормальному закону, равно параметру $a$, этого закона, т.е.
$$ M(X) = a, $$
а ее дисперсия — параметру $\sigma^2$, т.е.
$$ D(X) = \sigma^2 $$

Функция Лапласа:
$$ \Phi(x) = \frac{2}{\sqrt{2\pi}} \int_0^x e^{-\frac{t^2}{2}} {d}t = \frac{1}{\sqrt{2\pi}} \int_{-x}^x e^{-\frac{t^2}{2}} {d}t $$

Функция распределения случайной величины $X$, распределенной по нормальному закону распределения, выражается через функцию Лапласса $\Phi(x)$ по формуле:
$$ F_N(x) = \frac{1}{2} + \frac{1}{2} \Phi(\frac{x-a}{\sigma}) $$

### Правило трех сигм
Если случайная величина $X$ имеет нормальный закон распределения с параметрами $a$ и $\sigma^2$, то есть $N(a;\sigma^2)$, то практически достоверно, что его значения заключены в интервале $(a-3\sigma, \; a+3\sigma)$.
$$ P(|X - a| > 3\sigma) = 1 - P(|X - a| \leq 3\sigma) = 1 - 0,9973 = 0,0027 $$

## Логарифмически-нормальное распределение
Непрерывная случайная величина $X$ имеет логарифмически нормальное распределение, если ее логарифм подчинен нормальному закону.
$$ F(x) = P(X < x) = P(lnX < lnx) = \frac{1}{\sigma \sqrt{2\pi}} \int_{-\infty}^{lnx} e^{-\frac{(t - lna)^2}{2\sigma^2}} {d}t $$
$$ \phi(x) = \frac{1}{\sigma \sqrt{2\pi} x} e^{-\frac{(lnx - lna)^2}{2\sigma^2}} $$

## Распределения некоторых случайных величин, представляющих функции нормальных величин
### $\chi^2$-рапределение
**Распределением $\chi^2$ с $k$ степенями свободы** называется распределение суммы квадратов $k$ независимых случайных величин, распределенных по стандартному нормальному закону, т.е.
$$ \chi^2 = \sum_{i=1}^k Z_i^2, $$
где $Z_i \; (i=1,2,...,k)$ имеет нормальное распределение $N(0;1).$

![chi distribution density](data/chi_dist_dens.png "Chi distribution density")

### Распределение Стьюдента
**Распределением Стьюдента (t-распределением)** называется распределение случайной величины
$$ t = \frac{Z}{\sqrt{\frac{1}{k}\chi^2}}, $$
где $Z$ — случайная величина, распределенная по стандартному нормальному закону $N(0;1)$; \
$\chi^2$ — независимая от $Z$ случайная величина, имеющая $\chi^2$-распределение с $k$ степенями свободы.
